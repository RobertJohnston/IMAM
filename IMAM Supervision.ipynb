```python
# Convert siteid to number
df2['siteid'] = pd.to_numeric(df2.siteid, errors='coerce', downcast='signed')
# print df2.siteid
```

```python
# Remove NaN from siteid
df2 = df2.query('siteid==siteid').query('1<siteid<3999990999')
```

```python
# create db with only supervision staff
supervision_df = df2[df2['siteid'] <= 3799]
# supervision siteids range from 1 to 3799

# Sort data
supervision_df = supervision_df.sort_values(by='siteid')
supervision_df
```
